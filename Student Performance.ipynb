```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Expanded_data_with_more_features.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop("Unnamed: 0",axis=1)
df.head()
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df, x="Gender",color="red")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#from the above chart we have analysed that :
#the number of females in the data
#is more then males
```

```python
Gb=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(Gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(Gb,annot=True,cmap="coolwarm")
plt.show()
```

```python
#from the above chart we have concluded that the education of the parents have a good impact on their scores.
```

```python
gb=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb,annot=True)
plt.show()
```

```python
#from the above chart we have concluded that the marital status of the parents have not impact on their scores.
```

```python
groupa=df.loc[(df["EthnicGroup"] == "group A")].count()
groupb=df.loc[(df["EthnicGroup"] == "group B")].count()
groupc=df.loc[(df["EthnicGroup"] == "group C")].count()
groupd=df.loc[(df["EthnicGroup"] == "group D")].count()
groupe=df.loc[(df["EthnicGroup"] == "group E")].count() 

l=['group A' 'group B' 'group C' 'group D' 'group E']
mylist=[groupa["EthnicGroup"],groupb["EthnicGroup"],groupc["EthnicGroup"],groupd["EthnicGroup"],groupe["EthnicGroup"]]
plt.pie(mylist,autopct="%1.2f%%")
plt.title("Distribution Of Groups")
plt.show()
```
