```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Zomato-data-.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#CONVERT THE DATA TYPE COLUMN RATING
```

```python
def handle_rating(value):
    value=str(value).split('/')
    value=value[0]
    return float(value)

df["rate"]=df["rate"].apply(handle_rating)
print(df)
```

```python
df["rate"]=df["rate"].astype(float)
```

```python
#TYPE OF RESTAURENT
```

```python
sns.countplot(x=df["listed_in(type)"])
plt.show()
```

```python
#conclusion - majority of the restaurent falls in dinning category
```

```python
 groupby_data=df.groupby("listed_in(type)")["votes"].sum()
result=pd.DataFrame({"votes":groupby_data})
plt.plot(result,c="green",marker="o")
plt.xlabel("type of restaurent",c="red")
plt.ylabel("votes",c="red")
```

```python
#conclusion = dinning  restaurent has receive maximum vote
```
